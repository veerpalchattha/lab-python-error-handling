def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = initialize_inventory(products)
inventory
````

````python
product_items = []
total_price = float()
customer_orders = ('keychain', 'hat')

#function to calculate the total price of the customer order.
def totaPriceOrder():
    """
        Calculate the total price of the customer order

        Parameters: Empty.
    """
    for product in customer_orders:
        valid_input = False
        while not valid_input:
            try:
                price = float(input(f"Enter the price of the product {product}: "))
                if price >= 0.0:
                    product_items.append(float(price))
                    valid_input = True
                else:
                    print("Price cannot be negative. Please enter a valid Price.")   
            except ValueError:
                print("Invalid input. Please enter a valid Price.")
                
    price_sum = sum(float(price) for price in product_items)
    return price_sum
    
total_price = totaPriceOrder()
print(f"Total price: {total_price}")
````

````python
customer_orders = set()
    
def get_customer_orders(inventory):
    """
        Returns the customer orders set

        Parameters: inventory.
    """
    count = input("Enter the number of customer orders: ")
    if (count.isdigit() != True or count <= 0):
        print("Please enter a valid numeric quantity.")
        count = input("Enter the number of customer orders: ")
    for i in range(0, int(count)):
        valid_input = False
        while not valid_input:
            try:
                product = input(f"Enter the name for the product: ")
                if (product in inventory.keys() and inventory.get(product) != 0):
                    customer_orders.add(product)
                    valid_input = True
                else:
                    print("Please enter a valid product.")
            except KeyError:
                print("Invalid Product. Please enter a valid Price.")
                
get_customer_orders(inventory)
print("Customer orders are: ",customer_orders)
````
